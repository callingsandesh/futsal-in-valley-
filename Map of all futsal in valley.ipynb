```python
import openrouteservice
from openrouteservice import convert
```

```python
with open('api_key.txt','r') as file:
    api = file.readlines()
```

```python
client = openrouteservice.Client(key=api[0])
```

## SELECT YOUR LOCATION

```python
home = (27.671270,85.401617)
```

```python
max_distance = 10
```

```python
home
```

```python
list(home)
```

```python
map_home = folium.Map(location=list(home),zoom_start=15,min_zoom=11,max_zoom=17)
folium.Marker(list(home),popup='Home',icon=folium.Icon(color='red', icon='home', prefix='fa')).add_to(map_home)
```

```python
map_home
```

```python
dict_data={}
for item in points:
    #print(item)
    #set location coordinates in longitude,latitude order
    coords = ((home[1],home[0]),(item[1],item[0]))
    res = client.directions(coords)
    distance=round(res['routes'][0]['summary']['distance']/1000,1)
    if distance< max_distance:
        geometry = client.directions(coords)['routes'][0]['geometry']
        decoded = convert.decode_polyline(geometry)
        #duration_txt=str(round(res['routes'][0]['summary']['duration']/60,1))
        dict_data[(item[0],item[1])]=[distance , decoded,item[2],item[3]]
```

```python
for key,value in dict_data.items():
    #print(key)
    folium.Marker(list(key),popup=str(value[2])+"\n"+str(value[3])).add_to(map_home)
    folium.GeoJson(value[1]).add_child(folium.Popup(str(value[0])+" "+"KM",max_width=300)).add_to(map_home)
```
